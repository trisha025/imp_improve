```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('/Users/trishasinghal/Documents/table.png')
plt.figure(dpi=200)
plt.imshow(image)
sharpen_kernel = np.array([[-1,-1,-1], [-1,9,-1], [-1,-1,-1]])
sharpen = cv2.filter2D(image, -1, sharpen_kernel)

plt.figure(dpi=200)
plt.imshow(sharpen)

# cv2.imshow('sharpen', sharpen)
```

```python
from PIL import Image, ImageFilter, ImageEnhance
image = Image.open('/Users/trishasinghal/Documents/table.png')
sharpen_kernel = [
    -1, -1, -1,
    -1, 9, -1,
    -1, -1, -1
]
```

```python
# Create a kernel filter
kernel = ImageFilter.Kernel((3, 3), sharpen_kernel, scale=None, offset=0)

# Apply the sharpen filter
sharpened_image = image.filter(kernel)

enhancer = ImageEnhance.Sharpness(sharpened_image)
enhanced_image = enhancer.enhance(1.0)
```

```python
plt.figure(dpi=200)
plt.imshow(enhanced_image)
```
